```python
# 
yy = "2018"
rainfallpath = r"D:\Cornell\EthiopianDrought\Chirps2"
sifpath = r"D:\Cornell\EthiopianDrought\AData\SIF005Clip"
smpath = r"D:\Cornell\EthiopianDrought\AData\ESACCIV0.4.5"
etpath = r"D:\Cornell\EthiopianDrought\AData\ETClip"
evipath = r"D:\Cornell\EthiopianDrought\MOD13C2.006Ethiopia"
sifpath2 = r"D:\Cornell\EthiopianDrought\clipedSIf"

rainave =[]
rainaveall =[]
rainstd =[]
rainT = []

sifave =[]
sifaveall =[]
sifstd =[]
sifT =[]
sifave2 =[]
sifaveall2 =[]
sifstd2 =[]
sifT2 =[]



smave =[]
smstd=[]
smaveall =[]
smT =[]

etave = []
etstd =[]
etaveall = []
etT =[]
evistd =[]
eviave =[]
eviaveall =[]
eviT = []


for idx,month in enumerate(months):
    
    rainave.append(chirpsAveMap(rainfallpath ,yy, idx+1)[0])
    rainaveall.append(calChirpsAllAve(rainfallpath,idx+1)[0])
    rainstd.append((calChirpsAllAve(rainfallpath,idx+1)[2]))
    rainT.append(idx+1)
  
    sifave.append(sifAveMap(sifpath ,yy, idx+1)[0])
    sifaveall.append(calSifAllAve(sifpath,idx+1)[0])
    sifstd.append(calSifAllAve(sifpath,idx+1)[2])
    sifT.append(idx+1)
    
    sifave2.append(sifAveMap(sifpath2 ,yy, idx+1)[0])
    sifaveall2.append(calSifAllAve(sifpath2,idx+1)[0])
    sifstd2.append(calSifAllAve(sifpath2,idx+1)[2])
    sifT2.append(idx+1)
    
    eviave.append(eviAveMap(evipath ,yy, idx+1)[0])
    eviaveall.append(calEviAllAve(evipath,idx+1)[0])
    evistd.append(calEviAllAve(evipath,idx+1)[2])
    eviT.append(idx+1)
    
    smave.append(esacciAveMap(smpath ,yy, idx+1)[0])
    smaveall.append(calEsacciAllAve(smpath,idx+1)[0])
    smstd.append(calEsacciAllAve(smpath,idx+1)[2])
    smT.append(idx+1)
    
    etave.append(etAveMap(etpath ,yy, idx+1)[0])
    etaveall.append(calEtAllAve(etpath,idx+1)[0])
    etstd.append(calEtAllAve(etpath,idx+1)[0])
    etT.append(idx+1)
rainaveall = np.array(rainaveall)
rainstd = np.array(rainstd)

sifaveall =np.array(sifaveall)
sifstd =np.array(sifstd)

sifaveall2 =np.array(sifaveall2)
sifstd2 =np.array(sifstd2)

smstd=np.array(smstd)
smaveall =np.array(smaveall)



etstd =np.array(etstd)
etaveall =np.array(etaveall)

evistd =np.array(evistd)

eviaveall =np.array(eviaveall)


    
f,(ax1,ax2,ax3,ax4,ax5,ax6) = plt.subplots(6, figsize=(20,16))
plt.title("{} Precipitation Anomaly".format(yy), fontsize=20)
# f.figure(figsize=(20,16))
ax1.plot(rainT, rainave,"r-s",label="{}".format(yy))
print(list(zip(rainT,rainave)))
ax1.plot(rainT,rainaveall,label="multiyear average")

ax1.fill_between(rainT, rainaveall+rainstd,rainaveall- rainstd,color="lightgray")
ax1.set_title("Precipitation")
ax1.legend()

ax2.plot(smT, smave,"r-s",label="{}".format(yy))
ax2.plot(smT,smaveall,label="multiyear average")
ax2.fill_between(smT, smaveall+smstd,smaveall- smstd,color="lightgray")
ax2.set_title("Soil Moisture")
ax2.legend()

ax3.plot(etT, etave,"r-s",label="{}".format(yy))
ax3.plot(etT,etaveall,label="multiyear average")
ax3.fill_between(etT, etaveall+etstd, etaveall- etstd,color="lightgray")
ax3.set_title("ET")
ax3.legend()

ax4.plot(sifT, sifave,"r-s",label="{}".format(yy))
ax4.plot(sifT,sifaveall,label="multiyear average")
ax4.fill_between(sifT, sifaveall+sifstd,sifaveall- sifstd,color="lightgray")
ax4.set_title("SIF")
ax4.legend()

ax5.plot(sifT2, sifave2,"r-s",label="{}".format(yy))
ax5.plot(sifT2,sifaveall2,label="multiyear average")
ax5.fill_between(sifT2, sifaveall2+sifstd2,sifaveall2- sifstd2,color="lightgray")
ax5.set_title("SIF_Detrend")
ax5.legend()

ax6.plot(eviT, eviave,"r-s",label="{}".format(yy))
ax6.plot(eviT,eviaveall,label="multiyear average")
ax6.fill_between(eviT, eviaveall+evistd,eviaveall- evistd,color="lightgray")
ax6.set_title("EVI")

ax6.legend()

f.tight_layout()# 
plt.show()
```
